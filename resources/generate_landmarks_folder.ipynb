##Puntos de referencia faciales con dlib, OpenCV y Python

```python
args = {
	"shape_predictor": "/content/drive/MyDrive/Somnolencia/shape_predictor_68_face_landmarks.dat",
	"folder": "/content/drive/Othercomputers/Mi portátil/datasetSomnolencia/dormido",
	"image": "/content/drive/Othercomputers/Mi portátil/datasetSomnolencia/despierto/j5_0091.jpg"
}
```

```python
def detectar_puntos_faciales(img):
  # Inicializar el detector de rostros de dlib (basado en HOG)
  detector = dlib.get_frontal_face_detector()
  predictor = dlib.shape_predictor(args["shape_predictor"])

  # Convertir la imagen a escala de grises
  image = cv2.imread(img)

  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  # detectar rostros en la imagen en escala de grises
  rects = detector(gray, 1)

  # Inicializar una lista para almacenar los puntos de referencia faciales
  puntos_faciales = []

  # Bucle sobre las detecciones de rostros
  for (i, rect) in enumerate(rects):
    # Determinar los puntos faciales para la región del rostro
    shape = predictor(gray, rect)
    # Convertir los puntos faciales a un formato de NumPy array
    shape = face_utils.shape_to_np(shape)
    # Agregar los puntos de referencia faciales a la lista
    puntos_faciales.append(shape)

  return puntos_faciales
```

```python
def guardar_puntos_faciales(filename, puntos_faciales):
  with open(filename, 'w') as f:
    # Escribir metadatos del archivo
    f.write("version: 1\n")
    # Obtener la longitud total de los puntos faciales
    num_puntos = sum(len(arreglo) for arreglo in puntos_faciales)
    f.write(f"n_points: {num_puntos}\n")
    f.write("{\n")

    # Escribir las coordenadas de los puntos de referencia faciales
    for puntos in puntos_faciales:
        for (x, y) in puntos:
            f.write(f"{x} {y}\n")

    f.write("}\n")
```

```python
puntos_faciales =detectar_puntos_faciales(args["image"])
guardar_puntos_faciales("j5_0091.pts", puntos_faciales)
```

```python
def generar_archivos_pts_para_carpeta(folder):
  # Iterar sobre cada archivo de la carpeta
  for archivo in os.listdir(folder):
    if archivo.endswith(".jpg"):
      ruta_imagen = os.path.join(folder, archivo)
      puntos_faciales = detectar_puntos_faciales(ruta_imagen)
      ruta_pts = os.path.splitext(ruta_imagen)[0]+".pts"
      guardar_puntos_faciales(ruta_pts, puntos_faciales)

  print("Archivos generados con éxito")
```

```python

# Generar archivos .pts para cada imagen de la carpeta
generar_archivos_pts_para_carpeta(args["folder"])
```
